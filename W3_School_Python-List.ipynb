Loop Lists:

```python
# Using for loop to print all items individually

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
for x in favorite_fruits:
  print(x)
```

```python
# Using for loop to print all items individually (anather approach)

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
for i in range(len(favorite_fruits)):   #range() is creating a value (0, 5) range
  print(favorite_fruits[i])
```

```python
# Using a While Loop to print all items individually

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
i = 0
while i < len(favorite_fruits):
  print(favorite_fruits[i])
  i = i+1
```

```python
# Using list comprehension to print all items

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
[print(x) for x in favorite_fruits]
```

List Comprehension:

```python
# Using Loop to create new list from another list

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
names_n = []

for x in favorite_fruits:
  if "n" in x:
    names_n.append(x)

print(names_n)
```

```python
# Trying list comprehension instead of loop

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]

names_n = [x for x in favorite_fruits if "n" in x]

print(names_n)
```

```python
# Use of another condition in list

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]

names_n = [x for x in favorite_fruits if "mango" != x]

print(names_n)
```

```python
# List without condition

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]

names_n = [x for x in favorite_fruits]

print(names_n)
```

```python
# Creating list using range

names_n = [x for x in range(4)]

print(names_n)
```

```python
# Creating list using range and codition

numbers = [x for x in range (10) if x < 5]

print(numbers)
```

```python
# Trying to manipulate data before iteration by expression

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]

names_n = [x.title() for x in favorite_fruits]

print(names_n)
```

```python
# Replacing all input by hello using list

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]

names_n = ["hello" for x in favorite_fruits]

print(names_n)
```

```python
# Replacing poarticular Item/items by another using list

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]

names_n = [x if x != "mango" else "orange" for x in favorite_fruits]

print(names_n)
```

Sort List:

```python
# Sorting an alphabetic list
# Not defining order determine ascending order

favorite_fruits = ["apple", "banana", "orange", "kiwi", "mango"]

favorite_fruits.sort()

print(favorite_fruits)
```

```python
# Sorting a numeric list
# Not defining order determine ascending order

kg_fruits = [5, 15, 20, 100, 0, 6, 82]

kg_fruits.sort()

print(kg_fruits)
```

```python
# Sorting an alphabetic list in Descending order

favorite_fruits = ["apple", "banana", "orange", "kiwi", "mango"]

favorite_fruits.sort(reverse = True)

print(favorite_fruits)
```

```python
# Sorting a numeric list in Descending order

kg_fruits = [5, 15, 20, 100, 0, 6, 82]

kg_fruits.sort(reverse = True)

print(kg_fruits)
```

```python
# Customize Sort Function in list

#Creating a Function

def mynumber(n):
  return abs(n-40) # 'abs' for absolute value

kg_fruits = [5, 15, 20, 100, 0, 6, 82]

kg_fruits.sort(key = mynumber) # 'key' just make a criateria but don't change the value

print(kg_fruits)
```

```python
# Sorting an alphabetic list in Case sensitive mode

favorite_fruits = ["apple", "banana", "Orange", "kiwi", "mango"]

favorite_fruits.sort()

print(favorite_fruits)
```

```python
# Having no impact of case in an alphabetic list

favorite_fruits = ["apple", "banana", "Orange", "kiwi", "mango"]

favorite_fruits.sort(key = str.lower) # It is lower casing all value

print(favorite_fruits)
```

```python
# Reversing an alphabetic list

favorite_fruits = ["apple", "banana", "Orange", "kiwi", "mango"]

favorite_fruits.reverse()

print(favorite_fruits)
```

Join Lists:

```python
# Adding multiple lists (concatenation)

favorite_fruits = ["apple", "banana", "Orange", "kiwi", "mango"]
kg_fruits = [5, 15, 20, 100, 0, 6, 82]

all_data = favorite_fruits + kg_fruits

print(all_data)
```

```python
# Adding multiple lists (appending)
# It updates the initial list rather creating new one

favorite_fruits = ["apple", "banana", "Orange", "kiwi", "mango"]
kg_fruits = [5, 15, 20, 100, 0, 6, 82]

for x in kg_fruits:
  favorite_fruits.append(x)

print(favorite_fruits)
```

```python
# Adding multiple lists (extend)
# It updates the initial list rather creating new one

favorite_fruits = ["apple", "banana", "Orange", "kiwi", "mango"]
kg_fruits = [5, 15, 20, 100, 0, 6, 82]

favorite_fruits.extend(kg_fruits)

print(favorite_fruits)
```
