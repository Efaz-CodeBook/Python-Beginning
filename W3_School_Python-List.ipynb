Access List Items:

```python
# Printing one particular item from list

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
print(favorite_fruits[2])
```

```python
# Printing one particular item from list by negative indexing

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
print(favorite_fruits[-3])
```

```python
"""
Printing multiple items from list
In this case the items should consecutively placed in list
"""
# 1. when we need a spacific range (positive-approach)

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
print(favorite_fruits[1:4])
```

```python
# 2. when we need from beginning (positive-approach)

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
print(favorite_fruits[:4])
```

```python
# 3. when we need to end (positive-approach)

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
print(favorite_fruits[2:])
```

```python
# 4. when we need a spacific range (negative-approach)

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
print(favorite_fruits[-4:-2])
```

```python
# 5. when we need from beginning (positive-approach)

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
print(favorite_fruits[:-4])
```

```python
# 6. when we need to end (negative-approach)

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
print(favorite_fruits[-3:])
```

```python
# Trying if statement to find a value in a list

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]

if "kiwi" in favorite_fruits:
    print("Kiwi is one of my favorite fruits")
else:
    print("Kiwi is not my favorite fruit")
```

Change List Items:

```python
# Replace a value from list by position

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits[3] = "pineapple"
print(favorite_fruits)
```

```python
# Change range of values by position

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits[2:4] = "cherry", "grap"
print(favorite_fruits)
```

```python
# Change one value by two new values by position

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits[2:3] = "cherry", "grap"
print(favorite_fruits)
```

```python
# Change two value by two one values by position

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits[2:4] = ["cherry"] # Here square bracket is required to input
print(favorite_fruits)
```

```python
# To insert any value in any spacific position without replacing

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits.insert(2, "cherry")
print(favorite_fruits)
```

Add List Items:

```python
# Adding any item at the end of the list

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits.append("cherry")
print(favorite_fruits)
```

```python
# To insert any value in any spacific position

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits.insert(2, "grap")
print(favorite_fruits)
```

```python
# Add a group of items or a list into a list

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
more_fruits = ["pineapple", "watermelon", "jackfruit"]
favorite_fruits.extend(more_fruits)
print(favorite_fruits)
```

```python
# Add a group of items or iterable objects (tuple) into a list
# To add itemst it is not necessary to add list only

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
more_fruits = ("pineapple", "watermelon", "jackfruit")
favorite_fruits.extend(more_fruits)
print(favorite_fruits)
```

Remove List Items:

```python
# Removing any item from the list
# It removes only first item found in same name instead of all

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits.remove("mango")
print(favorite_fruits)
```

```python
# Removing any item from the list in specific position

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits.pop(4)
print(favorite_fruits)
```

```python
# No position value refers last item

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits.pop()
print(favorite_fruits)
```

```python
# Delet item from a list by del keyword

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
del favorite_fruits[4]
print(favorite_fruits)
```

```python
# Del keyword delets whole list if position is not declared

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
del favorite_fruits
print(favorite_fruits)
```

```python
# Clearing the list item to Nil

favorite_fruits = ["apple", "banana", "mango", "apple", "mango"]
favorite_fruits.clear()
print(favorite_fruits)
```
