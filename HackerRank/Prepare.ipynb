```python
def count_substring(string, sub_string):
    
    x = 0
   
    for i in range(0, len(string)):
        if string[i:i+len(sub_string)] == sub_string:
            x += 1
    return x


if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
    print(sub_string, string)
```

```python
s = "gss8g"

for i in range(len(s)):
    if i == len(s)-1:
        print(s[i].isalnum() == True)
        break
    elif s[i].isalnum() == True:
        print(True)
        break

for i in range(len(s)):
    if i == len(s)-1:
        print(s[i].isalpha() == True)
        break
    elif s[i].isalpha() == True:
        print(True)
        break

for i in range(len(s)):
    if i == len(s)-1:
        print(s[i].isdigit() == True)
        break
    elif s[i].isdigit() == True:
        print(True)
        break

for i in range(len(s)):
    if i == len(s)-1:
        print(s[i].islower() == True)
        break
    elif s[i].islower() == True:
        print(True)
        break

for i in range(len(s)):
    if i == len(s)-1:
        print(s[i].isupper() == True)
        break
    elif s[i].isupper() == True:
        print(True)
        break
```
