**3.1 What Is a List?**

```python
# A list is declared by square brackets ([])

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)
```

```python
# 3.1.1 Accessing Elements in a List

# Listd are ordered and it can be accessed by the positional value

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles[1])

# Title module can be used here to uppercase the first letter

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles[1].title())
```

```python
# 3.1.2 Index of a list

# Indexing of list starts from 0 instead of 1

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles[1]) # this shows the 2nd value in the list
print(bicycles[3]) # this shows the 4th value in the list
print(bicycles[0]) # this shows the 1st value in the list

# In python index can be counted from the last which starts from -1

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles[-1]) # this shows the last value in the list
print(bicycles[-3]) # this shows the 3rd last value in the list
```

```python
# 3.1.3 Using Individual Values from a List

# Concatenation of text and value of list

bicycles = ["runner", "alok", "duranta", "hilly"]
statement = "I love riding " + bicycles[2].title() + "."

print(statement)
```

**3.2 Changing, Adding, and Removing Element**

```python
# 3.2.1 Modifying Elements in a List
# Changing value of list

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)

bicycles[1] = "hero"

print(bicycles)
```

```python
# Adding Elements to a List

# Append module will add a new value at the last

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)

bicycles.append("hero")
print(bicycles)
```

```python
# Append is a very usefull module to create a list

motorbike = []  #an empty variable of list

motorbike.append("hero")
motorbike.append("yamaha")
motorbike.append("scarlet")

print(motorbike)
```

```python
# Inserting Elements into a List

# Insert can be used to add value in a particular position where other values just change position instead of replaced

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)

bicycles.insert(2, "hero") # (position index, value)
print(bicycles)
```

```python
# Removing Elements from a List

# Removing an Item Using the del Statement

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)

del bicycles[1]
print(bicycles)
```

```python
# Removing an Item Using the pop() Method

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)

popped_cycle = bicycles.pop() # pop item not only removes the value but also we can store the popped item

print(bicycles)
print(popped_cycle)
```

```python
# Printing the popped item with string while removing

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)

popped_cycle = bicycles.pop()

print(bicycles)
print("The last bicycles I owned was a " + popped_cycle.title() + ".")
```

```python
# Poping item from any position

bicycles = ["runner", "alok", "duranta", "hilly"]

popped_cycle = bicycles.pop(1)

print(bicycles)
print("The last bicycles I owned was a " + popped_cycle.title() + ".")
```

```python
# Removing an Item by Value

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)

bicycles.remove("duranta")

print(bicycles)
```

```python
# Remove module with a variable

bicycles = ["runner", "alok", "duranta", "hilly"]
print(bicycles)

cheap = "hilly"
bicycles.remove(cheap)
print(bicycles)
print("\nA " + cheap.title() + " is a very cheap product.")
```

**4.1 Looping Through an Entire List:**

```python
# Implementing for loop for a list

my_Guest = ["jishan", "ornob", "opu", "sayad"]

for guest in my_Guest:      # for item in list_of_items:
  print(guest)
```

```python
# Doing More Work Within a for Loop
# Printing whole message by loop

my_Guest = ["jishan", "ornob", "opu", "sayad"]

for guest in my_Guest:      # for item in list_of_items:
  print(f"{guest.title()} is one of my best friends")
```

```python
# Printing message by loop and use indentation

my_Guest = ["jishan", "ornob", "opu", "sayad"]

for guest in my_Guest:      # for item in list_of_items:
  print(f"{guest.title()} is one of my best friends")
  print(f"I love to chat with {guest.title()}.\n")
```

```python
# Doing Something After a for Loop
# To do some activity after loop we just need to break the indentation serial of loop

my_Guest = ["jishan", "ornob", "opu", "sayad"]

for guest in my_Guest:      # for item in list_of_items:
  print(f"{guest.title()} is one of my best friends")
  print(f"I love to chat with {guest.title()}.\n")

print(my_Guest) # Cutted the left spaces
```

```python
# if-else Statements
# else is required if the tested value is false

age = 15
if age >= 18:
    print("You are old enough to vote!")
    print("Have you registered to vote yet?")
else:
    print("Sorry, you are too young to vote.")
    print("Please register to vote as soon as you turn 18!")
```

```python
# Storing information and reduce text

# The if-elif-else Chain

mark = 50

if mark < 33:
    result = "Failed"
elif mark <80:
    result = "passed but not A+"
else:
    result = "ot A+"

print(f"You {result}.")
```

```python
# Using Multiple elif Blocks

mark = 50

if mark < 33:
    result = "Failed"
elif mark <60:
    result = "got B"
elif mark <80:
    result = "got A"
else:
    result = "ot A+"

print(f"You {result}.")
```

**5.3 Using if Statements with Lists:**

```python
# Checking for Special Items

favorite_fruits = ["apple", "mango", "pineapple"]

for fruits in favorite_fruits:
    print("Eating " + fruits +".")

print("\nFinished eating desert.")
```

```python
# Inserting if statement inside loop

favorite_fruits = ["apple", "mango", "pineapple"]

for fruits in favorite_fruits:
    if fruits == "mango":
        print("Stop! I have no mango today.")
    else:
        print("Eating " + fruits +".")

print("\nFinished eating desert.")
```

```python
# Checking That a List Is Not Empty
# If the list is epmty it will not run the for loop

favorite_fruits = []

if favorite_fruits:                 # Here favorite_fruits only indicates wheather it has value or not
    for fruits in favorite_fruits:
        print("Eating " + fruits +".")
    print("\nFinished eating desert.")
else:
    print("Today is a bad day for eating.")
```

```python
# Using Multiple Lists
# List to list search

favorite_fruits = ["kiwi", "apple", "mango", "pineapple", "banana"]

req_foods = ["apple", "mango", "lichi"]

for req_f in req_foods:
    if req_f in favorite_fruits:
        print("Eating the "+ req_f)
    else:
        print("Sorry, We don't have the fruit.")

print("\nFinished eating desert.")
```

```python
# 5-10. Checking Usernames:

cur_usernames = ["admin", "kabir", "ishan", "jishan", "kishan", "nishan"]

new_usernames = ["admin", "rabbi", "ishan", "Jishan", "fikir", "nishan"]

for new_user in new_usernames:
    if new_user.lower() in cur_usernames:
        print(f"This name {new_user} already exist in DB, Please insert new one.")
    else:
        print(f"This name {new_user} is available for you.")
```
