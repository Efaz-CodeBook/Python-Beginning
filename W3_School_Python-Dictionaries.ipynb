# W3 School: Python Dictionaries

* Dictionary stored data in pair, One is Key and another one is value (Key : 
Value)
* Dictionary is ordered, means it has positional index
* It is changable menas, we can directly apply modules to change data
* Dictionary doesn't allow duplicate value

```python
# Python Dictionaries: Beginning with structure

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic)
```

```python
# Accessing specific value by Keys

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic["model"])
```

```python
# Checking duplicate values in dictionary

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002,
    "year" : 1990     #Showing only this value
}

print(carDic)
```

```python
# Determine the dictionary length

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(len(carDic)) # Showing the number of Keys
```

```python
# Dictionary allows String, boolean, int, and list data types
# type() modules is allowed in dictionary

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : True,
    "year" : 2002,
    "color" : ["Red", "Black", "Mat"]
}

print(carDic)
print(type(carDic["brand"]))
print(type(carDic["model"]))
print(type(carDic["year"]))
print(type(carDic["color"]))
```

```python
# Dictionariescan be created by dict() constructor

carDic = dict(
    keys = "values",
    brand = "Porsche",
    model = "Cayenne",
    year = 2002
)

print(carDic)
```

Access Dictionary Items:

```python
"""
To access value of a dictionary, key name is reqired for specific data
As previous, to identify the position squre bracket is used where key name is used instead of number
"""

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic["year"])
```

```python
# get() module can also provide data by the key

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic.get("year"))
```

```python
# To get the list of keys, keys() method is used

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic.keys())
```

```python
# Trying to get the keys before and after update new keys

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic.keys()) # Before Adding new key

carDic["color"] = "red"

print(carDic.keys()) # After Adding new key
```

```python
# To get the values from dictionary using values() Method

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic.values())
```

```python
# To get the values from dictionary using values() method from before and after adding new

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic.values()) # Before Adding new value

carDic["color"] = "red"

print(carDic.values()) # # After Adding new value
```

```python
# To change the values from dictionary, need to use the particular key

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic) # Before updating new value

carDic["year"] = 1990

print(carDic) # # After updating new value
```

```python
# item() method stors both key and value in a single list and creats a tuple of lists

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic.items())
```

```python
# items() module used in dictionary to see the particular value of a keys
# Here, change will be displayed in list

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

print(carDic.items()) # Before updating new value

carDic["year"] = 1990

print(carDic.items()) # # After updating new value

# It can also happen for new keys

carDic["color"] = "red"

print(carDic.items()) # # After updating new value

#Trying a value search by condition

if "model" in carDic:
  print("Yes, the model of the car is " + carDic['model'])
```

Change Dictionary Items:

```python
# To change dictionary items, need to use the key

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}


carDic["year"] = 1992
print(carDic)
```

```python
# To Update Dictionary another module is used, update()

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

carDic.update({"year" : 1992})
print(carDic)
```

Add Dictionary Items:

```python
# To add items in dictionary a new key is added

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

carDic["color"] = "Blue"
print(carDic)
```

Remove Dictionary Items:

```python
# To remove items in dictionary need to remove by the key

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

carDic.pop("model")
print(carDic)
```

```python
# To remove a random item in dictionary popitem is used

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

carDic.popitem()
print(carDic)
```

```python
# Del method can also be used to delete a specific item

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

del carDic["model"]
print(carDic)
```

```python
# Del method can cause an error if the key is not defined as it delets whole dictionary

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

del carDic
print(carDic)
```

```python
# Clear method clear the all data into the dictionary

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

carDic.clear()
print(carDic)
```

Loop Dictionaries:

```python
# To print all dictionary item one by one loop is used

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

for x in carDic:
  print(x) # Here the key names will be printed
```

```python
# To print all dictionary values one by one the following is used

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

for x in carDic:
  print(carDic[x]) # Here the carDic[x] specifies that the x means the values of the dictionary
```

```python
# Anothe method "values()" can also use while defining x in loop

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

for x in carDic.values():
  print(x)
```

```python
# As before method "keys()" is also used while defining x in loop

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

for x in carDic.keys():
  print(x)
```

```python
# To print both keys and values the method items() is used

carDic = {
    "keys" : "values",
    "brand" : "Porsche",
    "model" : "Cayenne",
    "year" : 2002
}

for x, y in carDic.items():
  print(x, y)
```

Nested Dictionaries:

*   Nested dictionaries are dictionaries betweet dictionary

```python
carDic1 = {
    "brand" : "Porsche",
        "model" : "Cayenne",
        "year" : 2002
}
carDic2 = {
    "brand" : "Mercedes-Benz",
    "model" : "A-Class",
    "year" : 2020
}
carDic3 = {
    "brand" : "Toyota",
    "model" : "Corolla",
    "year" : 2022
}

myCars = {
    "carDic1" : carDic1,
    "carDic2" : carDic2,
    "carDic3" : carDic3
}

print(myCars)
```

```python
# Accessing Items in nested dictionaries is used by multiple square bracket[]

myCars = {
    "carDic1" : {
        "brand" : "Porsche",
        "model" : "Cayenne",
        "year" : 2002
    },
    "carDic2" : {
        "brand" : "Mercedes-Benz",
        "model" : "A-Class",
        "year" : 2020
    },
    "carDic3" : {
        "brand" : "Toyota",
        "model" : "Corolla",
        "year" : 2022
    }
}

print(myCars["carDic3"]["model"])
```

Dictionary Methods:

```python
# Trying some of the methods of Dictionary

myCars = {
    "carDic1" : {
        "brand" : "Porsche",
        "model" : "Cayenne",
        "year" : 2002
    },
    "carDic2" : {
        "brand" : "Mercedes-Benz",
        "model" : "A-Class",
        "year" : 2020
    },
    "carDic3" : {
        "brand" : "Toyota",
        "model" : "Corolla",
        "year" : 2022
    }
}

# clear() method inside nested dictionary
myCars["carDic1"].clear()

print(myCars)

# copy() method inside nested dictionary

carDic0 = myCars["carDic2"].copy()

print(carDic0)


# get() is used to get specific value by key

print(myCars.get("carDic3"))

# items() method is used to get key and value as tuples

carDic4 = myCars["carDic2"].items()

print(carDic4)

# keys() method is used to get keys

carDic4 = myCars["carDic2"].keys()

print(carDic4)

# values() method is used to get values

carDic4 = myCars["carDic2"].values()

print(carDic4)

# pop() method is used to delete items

myCars["carDic2"].pop("model")

print(myCars)

# popitems() method is used to delete last items

myCars["carDic3"].popitem()

print(myCars)
```

```python
# Dictionary setdefault() Method
"""
keyname : Required. The keyname of the item you want to return the value from
value	: Optional.
        If the key exist, this parameter has no effect.
        If the key does not exist, this value becomes the key's value
        Default value None
"""
carDic3 = {
    "brand" : "Toyota",
    "model" : "Corolla",
    "year" : 2022
}

x = carDic3.setdefault("model", "Axio")
y = carDic3.setdefault("model2", "Axio")

print(x)
print(y)
```

```python
# fromkeys() generates dictionary from two lists

x = ('brand', 'model', 'year')
y = 0

carDucs = dict.fromkeys(x,y)

print(carDucs)
```

```python
# update() is used to update value

carDic3 = {
    "brand" : "Toyota",
    "model" : "Corolla",
    "year" : 2022
}

carDic3.update({"model" : "Axio"})

print(carDic3)
```
